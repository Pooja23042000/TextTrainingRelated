```python
import pdb

num="86761016656"
rev=num[::-1]
print(rev)
i=0
res=""
while(i<len(rev)-1):
    pdb.set_trace()
    x=rev[i]+rev[i+1]
    print(x)
    if x=="32":
        res=res+ " "
        print(res)
    elif int(x) in range(65,91) or int(x) in range(97,100):
        res=res+chr(int(x))
        print(res)
    elif i+2<len(rev):
        x=x+rev[i+2]
        print(x)
        res=res+chr(int(x))
        print(res)
        i+=1
        print(i)
    i+=2
    print(i)
print(res)
```
