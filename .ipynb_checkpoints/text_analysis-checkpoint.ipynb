```python
# Convert dictionary into lists of definitions and corresponding CPC symbols
definitions = []
cpc_symbols = []
for symbol, definition in all_data.items():
    definitions.append(definition)
    cpc_symbols.append(symbol)
```

```python
def preprocess_text(texts):
    preprocessed_texts = []
    for text in texts:
        preprocessed_texts.append(text.lower())  # Convert to lowercase for uniformity
        text = re.sub(r'[^a-zA-Z\s]', '', text)
    return preprocessed_texts
```

```python
definitions= preprocess_text(definitions)
```

```python
def tokenize_text(texts, max_words):
    tokenizer = Tokenizer(num_words=max_words)
    tokenizer.fit_on_texts(texts)
    sequences = tokenizer.texts_to_sequences(texts)
    word_index = tokenizer.word_index
    return sequences, word_index
```

```python
max_words = 100000
max_seq_length = 1000
def_sequences, def_word_index = tokenize_text(definitions, max_words)
```

```python
def prepare_labels(labels):
    mlb = MultiLabelBinarizer()
    encoded_labels = mlb.fit_transform(labels)
    return encoded_labels, mlb.classes_
```

```python
encoded_labels, cpc_classes = prepare_labels(cpc_symbols)
```

```python
def_data = pad_sequences(def_sequences, maxlen=max_seq_length)
```

```python
def_train, def_test,labels_train,labels_test = train_test_split(
    def_data, encoded_labels,test_size=0.2, random_state=42
)
```

```python
embedding_dim = 100
lstm_units = 128

# Claim branch
def_input = Input(shape=(max_seq_length,))
def_embedding = Embedding(max_words, embedding_dim, input_length=max_seq_length)(def_input)
def_lstm = LSTM(lstm_units)(def_embedding)

# Output layer
output= Dense(len(cpc_classes), activation='sigmoid')(def_lstm)
```

```python
model = Model(inputs=def_input, outputs=output)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(def_train,labels_train,epochs=10, batch_size=64, validation_split=0.1)
```
